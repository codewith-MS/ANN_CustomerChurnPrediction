```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
##load the trained model
model = load_model('/content/drive/MyDrive/Udemy/ANN /model.h5')

#load the encoder and scaler
with open('/content/drive/MyDrive/Udemy/ANN /ANNohe.pkl','rb') as file:
  ANNohe = pickle.load(file)

with open('/content/drive/MyDrive/Udemy/ANN /ANNle.pkl','rb') as file:
  ANNle = pickle.load(file)

with open('/content/drive/MyDrive/Udemy/ANN /ANNscaler.pkl','rb') as file:
  ANNscaler = pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
# Encode Gender
with open('/content/drive/MyDrive/Udemy/ANN /ANNle.pkl', 'rb') as f:
    le = pickle.load(f)
input_df['Gender'] = le.transform(input_df['Gender'])
```

```python
with open('/content/drive/MyDrive/Udemy/ANN /ANNohe.pkl', 'rb') as f:
    ohe = pickle.load(f)
geo_encoded = ohe.transform(input_df[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=ohe.get_feature_names_out(['Geography']))
```

```python
# Combine
input_final = pd.concat([input_df.drop('Geography', axis=1), geo_encoded_df], axis=1)
```

```python
input_scaled = ANNscaler.transform(input_final)
prediction = model.predict(input_scaled)
print("Prediction:", prediction)
```

```python
#predict churn
prediction_proba = prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print('The customer is not likely to churn.')
```
